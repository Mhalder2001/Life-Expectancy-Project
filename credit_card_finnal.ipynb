## Building the supervised mode

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix,roc_auc_score,precision_score,f1_score,classification_report


```

```python
classifier = {
    "Logistic Regression":LogisticRegression(),
    "Decision Tree classifier":DecisionTreeClassifier(),
    "Random forest Regressor" :RandomForestClassifier(n_jobs=4, random_state=2024,verbose=False),
    "KNearest":KNeighborsClassifier(n_neighbors=5)
  

}
for name,clf in classifier.items():
    print(f"\n============={name}==========")
    clf.fit(x_train,y_train)
    y_predict=clf.predict(x_test)
    print(f"\n Accuracy:{accuracy_score(y_test,y_predict)}")
    print(f"\n Precision:{precision_score(y_test,y_predict)}")
    print(f"\n Recall:{recall_score(y_test,y_predict)}")
    print(f"\n F1 Score:{f1_score(y_test,y_predict)}")
    print(f"\n Confusion matrix:{confusion_matrix(y_test,y_predict)}")
    print(f"\n ROC AUC:{roc_auc_score(y_test,y_predict)}")
    print(f"\n Classification report:{classification_report(y_test,y_predict)}")
```


The output (41317, 199359) represents the number of transactions identified as potential frauds by the K-Means and DBSCAN clustering methods, respectively. Here's how you can interpret these numbers:

K-Means Clustering:

41317 transactions were identified as potential frauds by K-Means.
This number corresponds to the transactions that belong to the smaller cluster. In clustering, the assumption here is that fraudulent transactions might cluster together in a smaller group because they deviate from the norm.
DBSCAN Clustering:

199359 transactions were identified as potential frauds by DBSCAN.
In DBSCAN, transactions that do not fit well into any cluster (labeled as -1) are considered outliers. These outliers might represent fraudulent transactions, given that frauds often differ significantly from normal patterns.
Interpretation:
K-Means: The smaller cluster detected by K-Means suggests that there are 41,317 transactions that could potentially be fraudulent. K-Means might group transactions based on their similarities, and the smaller group is presumed to be fraud.

DBSCAN: DBSCAN identified a much larger number of outliers (199,359) as potential frauds. This is likely because DBSCAN is sensitive to noise and can classify many transactions as outliers if they do not fit into the main clusters.

Considerations:
High Number of Potential Frauds: The high number of transactions identified by DBSCAN might suggest that the parameters for DBSCAN (e.g., eps and min_samples) could be too strict, leading to a high number of outliers. You might want to adjust these parameters to reduce the number of false positives.

Comparison: The significant difference in the number of potential frauds between the two methods suggests that they are capturing different aspects of the data. It might be beneficial to combine insights from both methods or further refine the clustering parameters.
